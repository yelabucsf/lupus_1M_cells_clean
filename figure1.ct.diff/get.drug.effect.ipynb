```r
library(ggplot2)
library(reshape2)
cts=read.table('CLUESImmVar_processed.V6_joined_pivot_ct_perc.txt',sep = ',', header=T)

colnames(cts)[grep('now', colnames(cts))]
print(colnames(cts))
props=cts[, 14:25]
drugs=cts[,grep('now', colnames(cts))]
```

```r
##cellcept
cellcept.mat=NULL
for(i in 1:ncol(props)){
    print(colnames(props)[i])
    cellcept.mat=rbind(cellcept.mat, coef(summary(lm(props[, i] ~ drugs[, 2])))[2,])        
}

rownames(cellcept.mat)=colnames(props)
print(cellcept.mat)


```

```r
##plaquenil
plaq.mat=NULL
for(i in 1:ncol(props)){
    print(colnames(props)[i])
    plaq.mat=rbind(plaq.mat, coef(summary(lm(props[, i] ~ drugs[, 1])))[2,])        
}

rownames(plaq.mat)=colnames(props)
print(plaq.mat)

```

```r
##steroral
plaq.mat=NULL
for(i in 1:ncol(props)){
    print(colnames(props)[i])
    plaq.mat=rbind(plaq.mat, coef(summary(lm(props[, i] ~ drugs[, 3])))[2,])        
}

rownames(plaq.mat)=colnames(props)
print(plaq.mat)

```

```r
all=cbind(props, drugs)
df=melt(all, id.vars = c('plaquenilnow', 'cellceptnow', 'steroralnow'))
plt=ggplot(df,aes(x=as.factor(plaquenilnow), y=value, fill=variable)) + facet_grid(~variable) + geom_boxplot() + theme_bw()

print(plt)
pdf('supp.figures/plaquenil.composition.pdf')
print(plt)
dev.off()

plt=ggplot(df,aes(x=as.factor(cellceptnow), y=value, fill=variable)) + facet_grid(~variable) + geom_boxplot() + theme_bw()
print(plt)

pdf('supp.figures/cellcept.composition.pdf')
print(plt)
dev.off()

plt=ggplot(df,aes(x=as.factor(steroralnow), y=value, fill=variable)) + facet_grid(~variable) + geom_boxplot() + theme_bw()
print(plt)

pdf('supp.figures/steroral.composition.pdf')
print(plt)
dev.off()




```
